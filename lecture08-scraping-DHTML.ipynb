### webdriver 태그/ID/Class 사용하기

```python
driver.find_element_by_id("search_term").clear()
driver.find_element_by_id("search_term").send_keys("korea")
driver.find_element_by_id("search").click()
```

### webdriver CSS Selector 사용하기

```python
results = driver.find_element_by_id("results")
for tag in results.find_elements_by_tag_name("a"):
    print(tag.text)
    print(tag.get_attribute("href"))
```

```python
results = driver.find_elements_by_css_selector("#results a")
for tag in results:
    print(tag.text)
    print(tag.get_attribute("href"))
```

### webdriver XPath 사용하기

- . => 현재위치
- // => 자손
- / => 자식
- [@어트이브트="값"]

```python
div = driver.find_elements_by_xpath("//div[@id='results']")
div
```

```python
len(driver.find_elements_by_xpath("//div[@id='results']//a"))
```

```python
for tag in driver.find_elements_by_xpath("//div[@id='results']"):
    print(tag.get_attribute('href'))
    print(tag.text)
```
